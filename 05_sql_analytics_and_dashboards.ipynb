```python
"""# 05 SQL Analytics & Dashboards

This notebook contains business-focused SQL queries built on Gold tables.

These queries power dashboards for:
- Category performance
- Product insights
- Expiry risk monitoring
"""
```

```python
%sql
SELECT
  category,
  total_records AS total_items,
  ROUND(avg_price, 2) AS avg_price,
  near_expiry_count
FROM gold_category_metrics
ORDER BY total_items DESC;
```

```python
%sql
SELECT
  category,
  near_expiry_items,
  total_items,
  ROUND(expiry_risk_pct, 2) AS expiry_risk_pct
FROM gold_expiry_risk
ORDER BY expiry_risk_pct DESC;
```

```python
%sql
SELECT
  medicine_brand_name,
  manufacturing_company,
  category,
  batch_count,
  ROUND(avg_price, 2) AS avg_price
FROM gold_product_metrics
ORDER BY batch_count DESC
LIMIT 20;
```

```python
%sql
SELECT
  category,
  ROUND(MIN(avg_price), 2) AS min_price,
  ROUND(MAX(avg_price), 2) AS max_price,
  ROUND(AVG(avg_price), 2) AS mean_price
FROM gold_product_metrics
GROUP BY category
ORDER BY mean_price DESC;
```
